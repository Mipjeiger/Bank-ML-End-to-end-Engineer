# Understanding pipelines

### A pipeline is composed of a set of input parameters and a list of the steps in this workflow.

## Building pipelines using kubeflow
* ### Orchestrate repeatable ML workflows.
* ### Accelerate experimentation by running a workflow with different sets of hyperparameters.

## Purpose of building kubeflow pipelines heres:
* ### Orchestrates data pipelines for checking data integration
* ### Model evaluation checking in kubeflow as controller pipelines

## Design pipeline

```python
import warnings
warnings.filterwarnings('ignore')
import kfp
from kfp import dsl
import pandas as pd

@dsl.component(base_image='python:3.11')
def load_data(data_path: str) -> str:
    import pandas as pd
    df = pd.read_csv(data_path)
    print(f"✓ Loaded: {df.shape}")
    output_path = '/tmp/merged_data.csv.gz'
    df.to_csv(output_path, index=False, compression='gzip')
    return output_path

# Create pipeline
@dsl.pipeline(name='data-pipeline')
def data_pipeline():
    load_task = load_data(data_path='../../Database/data/eda_banking.csv.gz')

# Compile to YAML
kfp.compiler.Compiler().compile(
    pipeline_func=data_pipeline,
    package_path='data_pipeline.yaml'
)
print("✓ Pipeline compiled successfully!")
```
